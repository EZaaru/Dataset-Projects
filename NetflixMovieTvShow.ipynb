# Replace nan values with unknown


```python
df['director'].fillna('Unknown',inplace=True)
df['cast'].fillna('Unknown',inplace=True)
```

```python
df[df['country'].isna()].head()
```

```python
df['listed_in'].value_counts()
```

```python
mapping = {
    'Korean': 'South Korea',
    'Bollywood': 'India',
    'Anime': 'Japan',
    'Spanish': 'Spain',
    'Turkish': 'Turkey',
    'Italian': 'Italy',
    'French': 'France',
    'German': 'Germany',
    'Mexican': 'Mexico',
    'British': 'United Kingdom',
    'Indian': 'India'
}

def infer_country(row):
    if pd.isna(row['country']):
        listed = str(row['listed_in'])
        for key, val in mapping.items():
            if key.lower() in listed.lower():
                return val
        return 'Unknown Country'
    else:
        return row['country']

df['country'] = df.apply(infer_country, axis=1)
```

```python
df.dropna(subset=['date_added','rating'],inplace=True)
#drop rows have null in date_added , rating
# Basically It drops nan values from the given rows 
# if in data_added there is nan value it drops that nan value and continue further and same with rating 
```

# Feature Engineering


```python
# column show number of countries that the movie act at it
df['num_country']=df['country'].apply(lambda x:0 if x=='Unknown Country' else len(x.split(',')))

# column show number of cast who act in movie
df['num_cast']=df['cast'].apply(lambda x:0 if x=='Unknown' else len(x.split(',')))
```

```python
df['country'].value_counts()
```

```python
rating_groups = {
    'Kids': ['TV-Y', 'TV-Y7', 'TV-G', 'G', 'TV-Y7-FV'],
    'Teens': ['TV-PG', 'PG', 'PG-13', 'TV-14'],
    'Adults': ['TV-MA', 'R', 'NC-17', 'NR', 'UR']
}

# column show rating_category

def categorize_rating(rating):
   for key,val in rating_groups.items():
      if rating in val:
         return key

df['rating_category'] = df['rating'].transform(lambda x: categorize_rating(x))
```

```python
# Duration processing

df['duration_value']=df['duration'].apply(lambda x:int(x.split()[0]))
df['duration_type']=df['duration'].apply(lambda x:x.split()[1])
```

```python
df_movie=df[df['type']=='Movie']
df_tv=df[df['type']=='TV Show']

g_mov=df_movie.groupby('release_year')['title'].count()
g_mov=g_mov.sort_values(ascending=False)

g_tv=df_tv.groupby('release_year')['title'].count()
g_tv=g_tv.sort_values(ascending=False)
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,5))
ax[0].plot(g_mov.index,g_mov.values)
ax[1].plot(g_tv.index,g_tv.values)
ax[0].set_title('Movie distribution time')
ax[1].set_title('TV Show distribution time')
ax[0].set_xlabel('year')
ax[1].set_xlabel('year')
ax[0].set_ylabel('count')
ax[1].set_ylabel('count')
plt.show()
```

```python
fig,ax=plt.subplots(2,1,figsize=(20,20))
ax[0].bar(g_mov.index,g_mov.values)
ax[0].set_xlabel('genre')
ax[0].set_ylabel('count',rotation=90)
ax[0].set_title('Movie')
ax[0].tick_params(axis='x', rotation=90)
ax[1].bar(g_tv.index,g_tv.values)
ax[1].set_xlabel('genre')
ax[1].set_ylabel('count')
ax[1].set_title('TV Show')
ax[1].tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```
